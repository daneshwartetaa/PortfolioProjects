# Scrapping data from Wikipedia

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Public_sector_undertakings_in_India'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
soup.find_all('table')[1]
```

```python
soup.find('table', class_ = 'wikitable')
```

```python
table = soup.find_all('table')[1]
```

```python
print(table)
```

```python
CPSEs_titles = table.find_all('th')
```

```python
CPSEs_titles
```

```python
CPSEs_titles_table = [title.text.strip() for title in CPSEs_titles ][:4]

print(CPSEs_titles_table)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = CPSEs_titles_table)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:11]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
